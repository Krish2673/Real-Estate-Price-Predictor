```python
filtered_df['Posted_On']
```

```python
def convert_posted_on(value):
    try:
        num, unit, *_ = value.split()
        num = int(num)
        if 'day' in unit:
            return num
        elif 'week' in unit:
            return num * 7
        elif 'month' in unit:
            return num * 30
        elif 'year' in unit:
            return num * 365
    except:
        return np.nan

filtered_df['Posted_days_ago'] = filtered_df['Posted_On'].apply(convert_posted_on)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize encoder
le_sub = LabelEncoder()

# Fit & transform
filtered_df['Sub_urban_encoded'] = le_sub.fit_transform(filtered_df['Sub_urban_name'])

# (Optional) Drop the original
filtered_df.drop('Sub_urban_name', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Instantiate the scaler
scaler = StandardScaler()

# Apply scaling only to the selected numeric columns
cols_to_scale = ['Price_per_unit_area', 'Latitude', 'Longitude']
filtered_df[cols_to_scale] = scaler.fit_transform(filtered_df[cols_to_scale])
```
